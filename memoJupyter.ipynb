## Pour disposer d'un environnement interactif Jupyter :
### Local :
- La solution la plus pratique est d'installer une distribution Anaconda : https://www.anaconda.com/distribution/
 
### En ligne :

- Chez Microsoft Azure Notebooks : https://notebooks.azure.com/
    > *Le travail de Tony HIRST permet d'explorer : https://notebooks.azure.com/OUsefulInfo/projects/gettingstarted*

- Chez Google Colaboratory : https://colab.research.google.com
    > *Google a participé à l'origine de la création du projet Jupyter puis a poursuivi son développement indépendemment pour des besoins internes avant que de le rendre public en l'ajoutant à sa suite d'outils bureautique. C'est différent mais compatible et permet le partage...*

- Pour ouvrir de façon interactive mais temporaire un dépot GitHub ou GitLab dans un environnement Jupyter : https://mybinder.org/.
    > ***Attention**, il faut ajouter un [fichier de configuration](https://mybinder.readthedocs.io/en/latest/using.html#preparing-a-repository-for-binder) soit en [pip](https://github.com/binder-examples/requirements) soit en [conda](https://github.com/binder-examples/conda) pour indiquer à la machine virtuelle les paquets (Noyaux, Modules, Extensions) à installer.*

- Pour installer son propre serveur JupyterHub :
    - le plus simple, suivre la documentation : https://tljh.jupyter.org/en/latest/
> *[nbgitpuller]() pour distribuer des fichiers supports de cours aux élèves.*

    - autres solutions :
        - https://jupyterhub.readthedocs.io
        - https://z2jh.jupyter.org
        - https://pythonforundergradengineers.com/installing-jupyter-hub.html


## Extensions et/ou Noyaux complémentaires :
>*Certaines distributions viennent avec des noyaux (Kernels) et des extensions déjà intégrées.*

- Une liste assez exhaustive de solutions : https://awesomeopensource.com/project/markusschanta/awesome-jupyter

- Diaporama RISE : https://github.com/damianavila/RISE
    > [Tutoriel de Microsoft Azure Notebooks pour RISE](https://docs.microsoft.com/fr-fr/azure/notebooks/present-jupyter-notebooks-slideshow)

- [nbconvert](https://nbconvert.readthedocs.io) pour convertir les Notebook du format `.ipynb` vers `.pdf`, `.tex`,...

- [metakernel](https://github.com/Calysto/metakernel) pour inclure des codes Blockly, Processing, et bien plus encore...: [documentation](http://calysto.github.io/metakernel/), [exemples](http://nbviewer.jupyter.org/github/Calysto/metakernel/tree/master/examples/) : [Python Tutor](https://jupyter.brynmawr.edu/services/public/dblank/Examples/Tutor%20Magic%20in%20IPython.ipynb), ...

- [nbtutor](https://github.com/lgpage/nbtutor) pour visualiser l'exécution d'un code Python ligne par ligne dans un notebook tel qu'avec [Python Tutor](https://github.com/lgpage/nbtutor), lire [l'article de Tony Hirst](https://blog.ouseful.info/2017/01/27/python-code-stepper-debugger-tutor-for-jupyter-notebooks-nbtutor/) à ce sujet.

- [nbautoeval](https://github.com/parmentelat/nbautoeval)  une solution pour créer des exercices d'autoévaluation en python dans un Jupyter notebook.

- [Variable Inspector](https://jupyter-contrib-nbextensions.readthedocs.io/en/latest/nbextensions/varInspector/README.html) est contenue dans la collection [Unofficial Jupyter Notebook Extensions](https://github.com/ipython-contrib/jupyter_contrib_nbextensions) qui intègre notamment le gestionnaire d'extensions [jupyter_nbextensions_configurator](https://github.com/Jupyter-contrib/jupyter_nbextensions_configurator)

## Jupyterlab : 
> L'environnement Jupyter du futur !
>
> Pour basculer de l'environnement Jupyter NoteBook classique à l'environnement Lab, il suffit de remplacer dans l'URL le `tree` par un `lab` et inversement...
>
>C ’est une nouvelle génération de l’interface web utilisateur, plus évoluée, qui intègre les applications de Notebook et de Terminal classiques et adjoint une console Ipython et de nouvelles fonctionnalités graphiques pour la gestion des fichiers de l’espace personnel et l’affichage interactif simultané sur de multiples fenêtres.
>
> De plus, cet environnement autorise l’intégration d’applications en extensions comme un éditeur de diagrammes ou encore un éditeur/compilateur LaTeX…
